```python
dfb1 = dfb.loc[:, 'Average Rt(min)':'Average Mz']
dfb2 = dfb.loc[:, 'Box1-serum-pool1':'Box2-urine-pool4']
```

```python
def corr2_coeff_rowwise2(A,B):
    A_mA = A - A.mean(1)[:,None]
    B_mB = B - B.mean(1)[:,None]
    ssA = np.einsum('ij,ij->i',A_mA,A_mA)
    ssB = np.einsum('ij,ij->i',B_mB,B_mB)
    return np.einsum('ij,ij->i',A_mA,B_mB)/np.sqrt(ssA*ssB)
```

```python
df = pd.Series(corr2_coeff_rowwise2(dfa2.values,dfb2.values))
```

```python
df = pd.DataFrame(df)
df
```
