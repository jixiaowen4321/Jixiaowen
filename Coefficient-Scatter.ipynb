```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('CorPOS_GAdelivery_urine.csv', engine='c').drop(columns=['Unnamed: 0', 'sample_id'])
df2 = pd.read_csv('CorPOS_GAdelivery_serum.csv', engine='c').drop(columns=['Unnamed: 0', 'sample_id'])
```

```python
corr_matrix = df1.corr(method='spearman')['ga_delivery']
coef = list(corr_matrix)[:-1]
plt.scatter(x=list(df1.columns[:-1]), y=coef, s=0.8)
plt.gca().set_xticklabels([])
plt.gca().axes.get_xaxis().set_ticks([])
plt.savefig('urine-ga-scatter.png')
```

```python
df2L = np.log10(df2)
```

```python
corr_matrix = df2.corr(method='spearman')['ga_delivery']
coef = list(corr_matrix)[:-1]
```

```python
plt.scatter(x=list(df2L.columns[:-1]), y=coef, s=0.8)
plt.gca().set_xticklabels([])
plt.gca().axes.get_xaxis().set_ticks([])
plt.savefig('serum-ga-scatter.png')
```
